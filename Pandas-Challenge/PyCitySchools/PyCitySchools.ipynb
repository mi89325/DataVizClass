```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
total_schools = len(school_data_complete["school_name"].unique())
total_students = school_data_complete["Student ID"].count()
total_budget = school_data_complete["budget"].sum()
avg_math_score = school_data_complete["math_score"].mean()
avg_read_score = school_data_complete["reading_score"].mean()
avg_overall_score = (avg_math_score + avg_read_score)/2
pass_math_count = school_data_complete[(school_data_complete["math_score"] > 70)].count()["student_name"]
pass_math_percentage = (pass_math_count/total_students)*100
pass_read_count = school_data_complete[(school_data_complete["reading_score"] > 70)].count()["student_name"]
pass_read_percentage = (pass_read_count/total_students)*100
school_data_df = pd.DataFrame({"Total Schools":[total_schools], "Total Students":[total_students], "Total Budget":[total_budget],
                               "Average Math Score":[avg_math_score], "Average Reading Score":[avg_read_score], "Average Overall Score":[avg_overall_score],
                              "% Passing Math":[pass_math_percentage], "% Passing Reading":[pass_read_percentage]})
school_data_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
school_name = school_data.set_index(["school_name"])["type"]
total_amount_students = school_data_complete["school_name"].value_counts()
total_school_budget = school_data_complete["budget"].unique()
budget_per_student = (total_school_budget/total_amount_students)
avg_math_score_per_school = school_data_complete.groupby(["school_name"])["math_score"].mean()
avg_read_score_per_school = school_data_complete.groupby(["school_name"])["reading_score"].mean()
pass_math_df = school_data_complete[school_data_complete["math_score"] > 70]
pass_math = pass_math_df.groupby(["school_name"]).count()["student_name"]
percentage_pass_math = (pass_math/total_amount_students)*100
pass_read_df = school_data_complete[school_data_complete["reading_score"] > 70]
pass_read = pass_read_df.groupby(["school_name"]).count()["student_name"]
percentage_pass_read = (pass_read/total_amount_students)*100
overall_pass_percentage = (percentage_pass_read+percentage_pass_math)/2
school_summary_data_df = pd.DataFrame({"School Type":school_name, "Total Students":total_amount_students, "Total Budget":total_school_budget,
                                   "Budget Per Student":budget_per_student, "Math Score": avg_math_score_per_school, "Reading Score": avg_read_score_per_school,
                                   "Passing math %":percentage_pass_math, "Passing reading %":percentage_pass_read,"Passing Overall %":overall_pass_percentage})
school_summary_data_df.rename_axis("School Name")
```
